```python
import os
import torch
from torch.utils.data import Dataset, DataLoader
import pandas as pd

# Define hyperparameters
BATCH_SIZE = 64
EMBEDDING_DIM = 256
HIDDEN_DIM = 512
N_LAYERS = 2
DROPOUT = 0.5
LEARNING_RATE = 0.001
N_EPOCHS = 25
CLIP = 1.0
TEACHER_FORCING_RATIO = 0.5

# Custom Dataset class
class GrammarCorrectionDataset(Dataset):
    def __init__(self, input_tokens, target_tokens, input_word2idx, target_word2idx, max_len=50):
        self.input_tokens = input_tokens
        self.target_tokens = target_tokens
        self.input_word2idx = input_word2idx
        self.target_word2idx = target_word2idx
        self.max_len = max_len

    def __len__(self):
        return len(self.input_tokens)

    def __getitem__(self, idx):
        input_seq = self.input_tokens[idx]
        target_seq = self.target_tokens[idx]

        # Convert words to indices
        input_indices = [self.input_word2idx.get(word, self.input_word2idx['<UNK>']) for word in input_seq]

        # Add <SOS> at the beginning and <EOS> at the end for target sequence
        target_indices = [self.target_word2idx['<SOS>']] + \
                          [self.target_word2idx.get(word, self.target_word2idx['<UNK>']) for word in target_seq] + \
                          [self.target_word2idx['<EOS>']]

        # Pad sequences
        input_indices = input_indices[:self.max_len] + [self.input_word2idx['<PAD>']] * (self.max_len - len(input_indices))
        target_indices = target_indices[:self.max_len + 2] + [self.target_word2idx['<PAD>']] * (self.max_len + 2 - len(target_indices))

        return {
            'input': torch.tensor(input_indices, dtype=torch.long),
            'target': torch.tensor(target_indices, dtype=torch.long),
            'input_length': len(input_seq),
            'target_length': len(target_seq) + 2  # +2 for <SOS> and <EOS>
        }

# Create datasets
train_dataset = GrammarCorrectionDataset(train_input, train_target, input_word2idx, target_word2idx)
val_dataset = GrammarCorrectionDataset(val_input, val_target, input_word2idx, target_word2idx)

# Create data loaders
train_loader = DataLoader(train_dataset, batch_size=BATCH_SIZE, shuffle=True)
val_loader = DataLoader(val_dataset, batch_size=BATCH_SIZE)

# Check a sample from the dataset
sample = train_dataset[0]
print("Sample from dataset:")
print(f"Input tensor shape: {sample['input'].shape}")
print(f"Target tensor shape: {sample['target'].shape}")
print(f"Input length: {sample['input_length']}")
print(f"Target length: {sample['target_length']}")
```

```python
# Step 4: Define Encoder-Decoder Model Architecture
# Run this cell to define the model architecture

# Encoder Model
class EncoderRNN(nn.Module):
    def __init__(self, vocab_size, embedding_dim, hidden_dim, n_layers=1, dropout=0.5):
        super(EncoderRNN, self).__init__()
        self.hidden_dim = hidden_dim
        self.n_layers = n_layers

        self.embedding = nn.Embedding(vocab_size, embedding_dim)
        self.lstm = nn.LSTM(embedding_dim, hidden_dim, n_layers,
                           dropout=dropout if n_layers > 1 else 0,
                           batch_first=True, bidirectional=True)
        self.fc = nn.Linear(hidden_dim * 2, hidden_dim)
        self.dropout = nn.Dropout(dropout)

    def forward(self, x):
        # x shape: [batch_size, seq_len]
        embedded = self.dropout(self.embedding(x))
        # embedded shape: [batch_size, seq_len, embedding_dim]

        outputs, (hidden, cell) = self.lstm(embedded)
        # outputs shape: [batch_size, seq_len, hidden_dim * 2]
        # hidden shape: [n_layers * 2, batch_size, hidden_dim]
        # cell shape: [n_layers * 2, batch_size, hidden_dim]

        # Concatenate the final forward and backward hidden states
        hidden = torch.cat((hidden[-2,:,:], hidden[-1,:,:]), dim=1)
        cell = torch.cat((cell[-2,:,:], cell[-1,:,:]), dim=1)

        # Apply linear layer to combine directions
        hidden = self.fc(hidden)
        cell = self.fc(cell)

        # Reshape to match decoder expectations
        hidden = hidden.unsqueeze(0).repeat(self.n_layers, 1, 1)
        cell = cell.unsqueeze(0).repeat(self.n_layers, 1, 1)

        return outputs, (hidden, cell)

# Decoder Model
class DecoderRNN(nn.Module):
    def __init__(self, vocab_size, embedding_dim, hidden_dim, n_layers=1, dropout=0.5):
        super(DecoderRNN, self).__init__()
        self.vocab_size = vocab_size
        self.hidden_dim = hidden_dim
        self.n_layers = n_layers

        self.embedding = nn.Embedding(vocab_size, embedding_dim)
        self.lstm = nn.LSTM(embedding_dim, hidden_dim, n_layers,
                           dropout=dropout if n_layers > 1 else 0,
                           batch_first=True)
        self.fc_out = nn.Linear(hidden_dim, vocab_size)
        self.dropout = nn.Dropout(dropout)

    def forward(self, x, hidden, cell):
        # x shape: [batch_size, 1]
        x = x.unsqueeze(1)

        embedded = self.dropout(self.embedding(x))
        # embedded shape: [batch_size, 1, embedding_dim]

        output, (hidden, cell) = self.lstm(embedded, (hidden, cell))
        # output shape: [batch_size, 1, hidden_dim]

        prediction = self.fc_out(output.squeeze(1))
        # prediction shape: [batch_size, vocab_size]

        return prediction, hidden, cell

# Seq2Seq Model (Combining Encoder and Decoder)
class Seq2Seq(nn.Module):
    def __init__(self, encoder, decoder, device):
        super(Seq2Seq, self).__init__()
        self.encoder = encoder
        self.decoder = decoder
        self.device = device

    def forward(self, src, trg, teacher_forcing_ratio=0.5):
        # src shape: [batch_size, src_len]
        # trg shape: [batch_size, trg_len]

        batch_size = trg.shape[0]
        trg_len = trg.shape[1]
        trg_vocab_size = self.decoder.vocab_size

        # Tensor to store decoder outputs
        outputs = torch.zeros(batch_size, trg_len, trg_vocab_size).to(self.device)

        # Encode the source sequence
        _, (hidden, cell) = self.encoder(src)

        # First input to the decoder is the < SOS > token
        input = trg[:, 0]

        for t in range(1, trg_len):
            # Pass input, hidden state and cell state through decoder
            output, hidden, cell = self.decoder(input, hidden, cell)

            # Store output
            outputs[:, t] = output

            # Teacher forcing
            teacher_force = torch.rand(1).item() < teacher_forcing_ratio

            # Get the highest predicted token
            top1 = output.argmax(1)

            # Use teacher forcing or not
            input = trg[:, t] if teacher_force else top1

        return outputs

# Initialize model
encoder = EncoderRNN(len(input_word2idx), EMBEDDING_DIM, HIDDEN_DIM, N_LAYERS, DROPOUT).to(device)
decoder = DecoderRNN(len(target_word2idx), EMBEDDING_DIM, HIDDEN_DIM, N_LAYERS, DROPOUT).to(device)
model = Seq2Seq(encoder, decoder, device).to(device)

# Print model summary
print("Encoder:")
print(encoder)
print("\nDecoder:")
print(decoder)
print("\nTotal parameters:", sum(p.numel() for p in model.parameters()))
```

```python
# Step 5: Define Training and Evaluation Functions
# Run this cell to define training, evaluation, and metric calculation functions

# Initialize optimizer and criterion
optimizer = optim.Adam(model.parameters(), lr=LEARNING_RATE)
criterion = nn.CrossEntropyLoss(ignore_index=target_word2idx['<PAD>'])

# Training function
def train_epoch(model, dataloader, optimizer, criterion, clip, teacher_forcing_ratio):
    model.train()
    epoch_loss = 0

    for batch in tqdm(dataloader, desc="Training"):
        src = batch['input'].to(device)
        trg = batch['target'].to(device)

        optimizer.zero_grad()

        output = model(src, trg, teacher_forcing_ratio)

        # Reshape output and target for loss calculation
        output_dim = output.shape[-1]
        output = output[:, 1:].reshape(-1, output_dim)
        trg = trg[:, 1:].reshape(-1)

        # Calculate loss
        loss = criterion(output, trg)

        # Backpropagation
        loss.backward()

        # Gradient clipping
        torch.nn.utils.clip_grad_norm_(model.parameters(), clip)

        # Update parameters
        optimizer.step()

        epoch_loss += loss.item()

    return epoch_loss / len(dataloader)

# Evaluation function
def evaluate_epoch(model, dataloader, criterion):
    model.eval()
    epoch_loss = 0

    with torch.no_grad():
        for batch in tqdm(dataloader, desc="Evaluating"):
            src = batch['input'].to(device)
            trg = batch['target'].to(device)

            output = model(src, trg, 0)  # No teacher forcing during evaluation

            # Reshape output and target for loss calculation
            output_dim = output.shape[-1]
            output = output[:, 1:].reshape(-1, output_dim)
            trg = trg[:, 1:].reshape(-1)

            # Calculate loss
            loss = criterion(output, trg)

            epoch_loss += loss.item()

    return epoch_loss / len(dataloader)

# Function to calculate BLEU score
def calculate_bleu(model, dataset, target_idx2word, target_word2idx, max_len=50):
    smoothie = SmoothingFunction().method4
    references = []
    candidates = []

    model.eval()
    with torch.no_grad():
        for i in range(len(dataset)):
            batch = dataset[i]
            src = batch['input'].unsqueeze(0).to(device)

            _, (hidden, cell) = model.encoder(src)
            input_token = torch.tensor([target_word2idx['<SOS>']]).to(device)
            predicted_sentence = []

            for t in range(max_len):
                output, hidden, cell = model.decoder(input_token, hidden, cell)
                pred_token = output.argmax(1).item()

                if pred_token == target_word2idx['<EOS>']:
                    break
                if pred_token != target_word2idx['<PAD>'] and pred_token != target_word2idx['<SOS>']:
                    predicted_sentence.append(target_idx2word[pred_token])

                input_token = torch.tensor([pred_token]).to(device)

            # Convert to format required by BLEU
            candidate = predicted_sentence
            target_sentence = []
            trg = batch['target']
            for t in range(1, batch['target_length']):
                idx = trg[t].item()
                if idx == target_word2idx['<EOS>']:
                    break
                if idx != target_word2idx['<PAD>']:
                    target_sentence.append(target_idx2word[idx])

            candidates.append(candidate)
            references.append([target_sentence])

    bleu1 = np.mean([sentence_bleu(r, c, weights=(1, 0, 0, 0), smoothing_function=smoothie) for r, c in zip(references, candidates)])
    bleu2 = np.mean([sentence_bleu(r, c, weights=(0.5, 0.5, 0, 0), smoothing_function=smoothie) for r, c in zip(references, candidates)])
    bleu3 = np.mean([sentence_bleu(r, c, weights=(0.33, 0.33, 0.33, 0), smoothing_function=smoothie) for r, c in zip(references, candidates)])
    bleu4 = np.mean([sentence_bleu(r, c, weights=(0.25, 0.25, 0.25, 0.25), smoothing_function=smoothie) for r, c in zip(references, candidates)])

    return bleu1, bleu2, bleu3, bleu4, references, candidates

# Function to calculate ROUGE score
def calculate_rouge(references, candidates):
    rouge = Rouge()

    # Convert reference and candidate lists to proper format for ROUGE
    formatted_refs = [' '.join(ref[0]) for ref in references]
    formatted_cands = [' '.join(cand) if cand else '' for cand in candidates]

    # Make sure there are no empty candidates
    valid_refs = []
    valid_cands = []
    for ref, cand in zip(formatted_refs, formatted_cands):
        if cand:  # Only include non-empty candidates
            valid_refs.append(ref)
            valid_cands.append(cand)

    if not valid_cands:
        return {'rouge-1': {'f': 0.0}, 'rouge-2': {'f': 0.0}, 'rouge-l': {'f': 0.0}}

    try:
        scores = rouge.get_scores(valid_cands, valid_refs, avg=True)
        return scores
    except Exception as e:
        print(f"Error calculating ROUGE: {e}")
        return {'rouge-1': {'f': 0.0}, 'rouge-2': {'f': 0.0}, 'rouge-l': {'f': 0.0}}

print("Training and evaluation functions defined")
```

```python
# Step 6: Train the Model
# Run this cell to train the model for multiple epochs

# Training and evaluation
train_losses = []
valid_losses = []
best_valid_loss = float('inf')

start_time = time.time()

for epoch in range(N_EPOCHS):
    print(f"Epoch {epoch+1}/{N_EPOCHS}")

    # Training
    train_loss = train_epoch(model, train_loader, optimizer, criterion, CLIP, TEACHER_FORCING_RATIO)

    # Evaluation
    valid_loss = evaluate_epoch(model, val_loader, criterion)

    # Save losses for plotting
    train_losses.append(train_loss)
    valid_losses.append(valid_loss)

    # Save the best model
    if valid_loss < best_valid_loss:
        best_valid_loss = valid_loss
        torch.save(model.state_dict(), 'grammar_correction_no_attention.pt')
        print("Model saved!")

    print(f"Train Loss: {train_loss:.4f}, Validation Loss: {valid_loss:.4f}")

training_time = time.time() - start_time
print(f"Training completed in {training_time:.2f} seconds")

# Plot losses
plt.figure(figsize=(10, 6))
plt.plot(train_losses, label='Training Loss')
plt.plot(valid_losses, label='Validation Loss')
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.title('Training and Validation Loss')
plt.legend()
plt.grid(True)
plt.savefig('loss_plot_no_attention.png')
plt.show()

# Load the best model
model.load_state_dict(torch.load('grammar_correction_no_attention.pt'))
print("Best model loaded")
```

```python
# CHUNK 5: Define the Bahdanau Attention Layer
class BahdanauAttention(tf.keras.layers.Layer):
    def __init__(self, units):
        super(BahdanauAttention, self).__init__()
        # Dense layers to learn the attention mechanism
        self.W1 = tf.keras.layers.Dense(units)  # To process encoder output
        self.W2 = tf.keras.layers.Dense(units)  # To process the decoder hidden state
        self.V = tf.keras.layers.Dense(1)      # To compute attention score

    def call(self, query, values):
        """
        query: Decoder hidden state, shape == (batch_size, hidden_size)
        values: Encoder output, shape == (batch_size, max_len, hidden_size)

        Returns:
        context_vector: The context vector, shape == (batch_size, hidden_size)
        attention_weights: The attention weights, shape == (batch_size, max_len, 1)
        """
        # Add an extra axis to query so that it can be broadcasted with the values
        hidden_with_time_axis = tf.expand_dims(query, 1)  # Shape: (batch_size, 1, hidden_size)

        # Calculate attention score (Score shape: (batch_size, max_len, 1))
        score = self.V(tf.nn.tanh(self.W1(values) + self.W2(hidden_with_time_axis)))

        # Apply softmax to get attention weights (shape: (batch_size, max_len, 1))
        attention_weights = tf.nn.softmax(score, axis=1)

        # Multiply attention weights with the values to get the context vector
        context_vector = attention_weights * values  # Shape: (batch_size, max_len, hidden_size)

        # Sum across the sequence length dimension to get the final context vector
        context_vector = tf.reduce_sum(context_vector, axis=1)  # Shape: (batch_size, hidden_size)

        return context_vector, attention_weights
```

```python
from tensorflow.keras.layers import Lambda

# CHUNK 6: Define the model with attention
embedding_dim = 256
units = 512
dropout_rate = 0.2

# Encoder
encoder_inputs = Input(shape=(max_input_length,))
encoder_embedding = Embedding(input_vocab_size, embedding_dim)(encoder_inputs)
encoder_dropout = Dropout(dropout_rate)(encoder_embedding)
encoder_outputs, state_h, state_c = LSTM(units, return_sequences=True, return_state=True)(encoder_dropout)
encoder_states = [state_h, state_c]

# Decoder with Attention
decoder_inputs = Input(shape=(max_target_length,))
decoder_embedding = Embedding(target_vocab_size, embedding_dim)(decoder_inputs)
decoder_dropout = Dropout(dropout_rate)(decoder_embedding)

# Initialize attention mechanism
attention_layer = BahdanauAttention(units)

# Initialize decoder LSTM and dense layer
decoder_lstm = LSTM(units, return_sequences=True, return_state=True)
decoder_dense = Dense(target_vocab_size, activation='softmax')

# Initial decoder states
decoder_state_h, decoder_state_c = encoder_states

# Initialize an empty list to store all outputs
all_outputs = []

# Process each time step of the decoder
for t in range(max_target_length):
    # Get the context vector using attention
    context_vector, attention_weights = attention_layer(decoder_state_h, encoder_outputs)

    # Extract the input for this time step
    decoder_input_t = decoder_dropout[:, t:t+1]

    # Run the decoder LSTM cell for this time step
    decoder_output, decoder_state_h, decoder_state_c = decoder_lstm(
        decoder_input_t, initial_state=[decoder_state_h, decoder_state_c])

    # Use a Lambda layer to squeeze the output (remove unnecessary dimensions)
    decoder_output_squeezed = Lambda(lambda x: tf.squeeze(x, axis=1))(decoder_output)

    # Wrap the concatenation in a Lambda layer
    decoder_combined_context = Lambda(lambda x: tf.concat(x, axis=1))([context_vector, decoder_output_squeezed])

    # Pass through a dense layer to get the output for this time step
    output = decoder_dense(decoder_combined_context)

    # Append the output to the list
    all_outputs.append(output)

# Use a Lambda layer to stack all outputs along the time dimension
decoder_outputs = Lambda(lambda x: tf.stack(x, axis=1))(all_outputs)

# Define the model
model = Model([encoder_inputs, decoder_inputs], decoder_outputs)

# Compile the model
model.compile(
    optimizer='adam',
    loss='sparse_categorical_crossentropy',
    metrics=['accuracy']
)

# Model summary
model.summary()
```

```python
from tensorflow.keras.models import Model
from tensorflow.keras.layers import Input, LSTM, Embedding, Dense, Attention, Concatenate, RepeatVector
from tensorflow.keras.callbacks import ModelCheckpoint
from tensorflow.keras.utils import to_categorical
import time
import matplotlib.pyplot as plt

# Define the model architecture (simplified version for illustration)
def create_model(input_vocab_size, output_vocab_size, embedding_dim, hidden_units, max_input_len, max_output_len):
    # Encoder
    encoder_input = Input(shape=(max_input_len,))
    encoder_embedding = Embedding(input_dim=input_vocab_size, output_dim=embedding_dim)(encoder_input)
    encoder_lstm = LSTM(hidden_units, return_state=True)
    encoder_outputs, state_h, state_c = encoder_lstm(encoder_embedding)

    # Repeat the encoder output for each time step in the decoder sequence
    encoder_output_repeat = RepeatVector(max_output_len - 1)(encoder_outputs)  # Repeat for the length of the decoder sequence

    # Decoder with Attention
    decoder_input = Input(shape=(max_output_len-1,))  # Teacher forcing data excludes the last token
    decoder_embedding = Embedding(input_dim=output_vocab_size, output_dim=embedding_dim)(decoder_input)

    # Project the decoder embedding to match the encoder's output dimensionality
    decoder_projection = Dense(hidden_units)(decoder_embedding)  # Match dimensions with encoder output

    # Attention mechanism
    attention = Attention()([decoder_projection, encoder_output_repeat])  # Now both inputs have compatible shapes
    decoder_concat_input = Concatenate(axis=-1)([decoder_projection, attention])

    # Decoder LSTM
    decoder_lstm = LSTM(hidden_units, return_sequences=True)(decoder_concat_input)

    # Output layer
    decoder_output = Dense(output_vocab_size, activation='softmax')(decoder_lstm)

    # Define the model
    model = Model(inputs=[encoder_input, decoder_input], outputs=decoder_output)

    # Compile the model
    model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])

    return model

# Define input and output vocab sizes, embedding dimension, and hidden units
input_vocab_size = len(input_tokenizer.word_index) + 1  # Add 1 for padding token
output_vocab_size = len(target_tokenizer.word_index) + 1  # Add 1 for padding token
embedding_dim = 256
hidden_units = 512
max_input_len = X_train.shape[1]
max_output_len = y_train.shape[1]

# Create model
model = create_model(input_vocab_size, output_vocab_size, embedding_dim, hidden_units, max_input_len, max_output_len)

# Create teacher forcing data and apply one-hot encoding
def create_teacher_forcing_data(X, y, output_vocab_size):
    decoder_input_data = y[:, :-1]  # Exclude the last token
    decoder_target_data = y[:, 1:]  # Exclude the first token (startseq)

    # One-hot encode the target labels
    decoder_target_data = to_categorical(decoder_target_data, num_classes=output_vocab_size)

    return [X, decoder_input_data], decoder_target_data

# Prepare teacher forcing data
train_inputs, train_targets = create_teacher_forcing_data(X_train, y_train, output_vocab_size)
val_inputs, val_targets = create_teacher_forcing_data(X_val, y_val, output_vocab_size)

# Define callbacks
model_checkpoint = ModelCheckpoint(
    'grammar_correction_attention_model.h5',
    save_best_only=True,
    monitor='val_loss',
    verbose=1
)
```

```python
import numpy as np

# Ensure the reverse mapping exists
reverse_target_token_index = {index: word for word, index in target_tokenizer.word_index.items()}

# Function to decode sequence
def decode_sequence(input_seq):
    # Ensure input_seq is in the correct shape for the encoder
    input_seq = np.array(input_seq)

    # Encode the input as state vectors
    encoder_outputs, state_h, state_c = encoder_model.predict(input_seq)

    # Start with the start token (e.g., 'startseq' in the tokenizer)
    target_seq = np.zeros((1, 1))  # Shape: (1, 1), one token at a time
    target_seq[0, 0] = target_tokenizer.word_index.get('startseq', 0)

    # Output sequence and attention weights
    output_text = []
    attention_plot = []

    stop_condition = False
    while not stop_condition:
        # Predict the next token in the sequence
        output_tokens, h, c, attention_weights = decoder_model.predict(
            [target_seq, encoder_outputs, state_h, state_c])

        # Sample a token (taking the argmax of the predicted tokens)
        sampled_token_index = np.argmax(output_tokens[0])
        sampled_word = reverse_target_token_index.get(sampled_token_index)

        # Stop if 'endseq' is generated or if the sequence is too long
        if sampled_word == 'endseq' or len(output_text) > max_target_length:
            stop_condition = True
        else:
            # Append the predicted word or an empty string if None
            output_text.append(sampled_word if sampled_word is not None else '')
            attention_plot.append(attention_weights)

        # Update the target sequence for the next timestep
        target_seq = np.zeros((1, 1))  # Shape: (1, 1)
        target_seq[0, 0] = sampled_token_index

        # Update the decoder's states for the next timestep
        state_h, state_c = h, c

    # Ensure no None values are in the output text
    output_text = [word if word is not None else '' for word in output_text]

    # Convert list to a single string and return along with attention plot
    return ' '.join(output_text), np.array(attention_plot)
```
